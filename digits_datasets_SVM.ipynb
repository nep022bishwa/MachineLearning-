```python
X=df.drop(columns=['target'],axis=1)
y=df.target
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.svm import SVC
```

```python
model=SVC()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
#for linear karnel with low regularization parameter C=1
model_linear=SVC(C=1,kernel='linear')
model_linear.fit(X_train,y_train)
model_linear.score(X_test,y_test)
```

```python
#for poly kernel with High regularization parameter C=10
model_poly=SVC(C=10,kernel='poly')
model_poly.fit(X_train,y_train)
model_poly.score(X_test,y_test)
```

```python
#for poly kernel with High regularization parameter C=10
model_poly=SVC(C=10,kernel='poly',gamma='auto')
model_poly.fit(X_train,y_train)
model_poly.score(X_test,y_test)
```
